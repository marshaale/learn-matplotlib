```python
x = [1,2,3,4,5,6,7]
y = [28,27,31,33,35,29,26]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth=2.5)
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,'go--')
```

```python
plt.plot(x,y,marker='D',linestyle="dotted",alpha=0.5)
```

```python
days = np.arange(1,8)
max_t = np.random.randint(30,47,size=7)
min_t = np.random.randint(19,29,size=7)
avg_t = np.random.randint(27,33,size=7)
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,'bo--',label="Max",)
plt.plot(days,min_t,'go--',label="Min")
plt.plot(days,avg_t,'ro--',label="Avg")

plt.legend(loc="best")
plt.grid()
```
