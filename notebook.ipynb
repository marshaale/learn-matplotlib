```python
days = np.arange(1,8)
max_t = np.random.randint(30,47,size=7)
min_t = np.random.randint(19,29,size=7)
avg_t = np.random.randint(27,33,size=7)
```

```python
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t,'bo--',label="Max",)
plt.plot(days,min_t,'go--',label="Min")
plt.plot(days,avg_t,'ro--',label="Avg")

plt.legend(loc="best")
plt.grid()
```

```python
companies = ['Google','Facebook','Amazon']
revenues = [120,90,85]
```

```python
xpos = np.arange(len(companies))
xpos
```

```python
plt.title("Stock")
plt.ylabel("Revenue")
plt.bar(xpos,revenues)
plt.xticks(xpos,companies)
```

```python
profits = [20,30,18]
plt.title("Stock")
plt.ylabel("Revenue")
plt.bar(xpos+0.2,revenues,width=0.4,label='Revenue')
plt.bar(xpos-0.2,profits,width=0.4,label='Profit')
plt.xticks(xpos,companies)
plt.legend()
```

```python
plt.title("Stock")
plt.yticks(xpos,companies)
plt.xlabel("Revenue")
plt.barh(xpos-0.2,revenues,label="Revenue")
plt.barh(xpos+0.2,profits,label="Profit")
plt.grid()
plt.legend()
```
