```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("clean_cafe_sales.csv")
df.drop(columns=["Unnamed: 0"],inplace=True)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
payment_methods = df['Payment Method'].value_counts()

plt.title("Payment Methods")
plt.pie(payment_methods, labels=payment_methods.index.tolist(), autopct="%0.2f%%", explode=[0, 0.1, 0])
plt.axis('equal')
plt.show()
```

```python
item_totals = df[['Item','Total Spent']].groupby('Item')['Total Spent'].sum()
x = item_totals.index.tolist()
y = item_totals.tolist()
plt.bar(x,y)
plt.xlabel('Item')
plt.ylabel('Total Spent')
plt.grid()
plt.show()
```

```python
plt.title("Payment Methods")
plt.ylabel("Transactions")
plt.plot(payment_methods.index.tolist(),payment_methods,'bo--')
plt.show()
```

```python
plt.xlabel('Item quantity')
plt.hist(df['Quantity'])
plt.show()
```

```python
transaction_date = df['Transaction Date'].astype('datetime64[us]').value_counts()
plt.plot(transaction_date.index.tolist(),transaction_date,'bo--')
plt.ylabel('Transactions')
plt.xlabel('Date')
plt.show()
```

```python
plt.title('Daily Transactions')
plt.hist(transaction_date)
plt.show()
```

```python
revenues = df[['Location','Total Spent']].groupby('Location')['Total Spent'].sum()
plt.title("Total revenue across locations")
plt.pie(revenues,labels=revenues.index.tolist(),autopct="%0.2f%%",explode=[0,0.1])
plt.show()
```
